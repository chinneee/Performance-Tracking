```python
import pandas as pd
import os
import sys
import re

import gspread
from gspread_dataframe import set_with_dataframe
from oauth2client.service_account import ServiceAccountCredentials
from google.oauth2.service_account import Credentials
```

```python
# Đường dẫn đến thư mục chứa các file CSV
folder_path = '/Users/ppcintern/Desktop/Performance-Tracking/ADS_Daily'  # <-- Bạn cần thay đường dẫn này cho đúng

# Danh sách để lưu các DataFrame
all_dfs = []

# Duyệt qua từng file trong thư mục
for filename in os.listdir(folder_path):
    if filename.endswith(".csv"):
        # Regex để lấy phần date: chuỗi 8 chữ số (YYYYMMDD)
        match = re.search(r'_(\d{8})_', filename)
        if match:
            date_str = match.group(1)
            full_path = os.path.join(folder_path, filename)
            
            # Đọc file CSV
            df = pd.read_csv(full_path)
            
            # Thêm cột Date
            df["Date"] = pd.to_datetime(date_str, format="%Y%m%d")
            
            # Thêm vào danh sách
            all_dfs.append(df)

# Gộp tất cả lại thành 1 DataFrame
merged_df = pd.concat(all_dfs, ignore_index=True)

# Hiển thị 5 dòng đầu tiên
merged_df.tail()
```
